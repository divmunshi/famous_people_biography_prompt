```python
import PyPDF2

def extract_text_from_pdf(file_path, page_num=0):
    with open(file_path, 'rb') as file:
        pdf = PyPDF2.PdfReader(file)
        text = ''
        page = pdf.pages[page_num]
        text += page.extract_text()
        return text

# Example usage
pdf_file_path = './sample_data/franz-mehring-karl-marx-the-story-of-his-life-1918.compressed.pdf'
pdf_text = extract_text_from_pdf(pdf_file_path, 29)
print(pdf_text)
```

```python
import csv

# Split the input string into lines
lines = response.split('\n')

# Remove empty lines
lines = [line for line in lines if line.strip()]

# Split the header line to extract column names
header = lines[0].split(' | ')

# Split the remaining lines to extract data
data = [line.split(' | ') for line in lines[2:]]

# Save as CSV
filename = './sample_data/output.csv'
with open(filename, mode='a', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(data)

print(f"CSV file '{filename}' saved successfully.")
```

```python
def appendToCharcterOutput(text):
    try:
        # Split the input string into lines
        lines = text.split('\n')

        # Remove empty lines
        lines = [line for line in lines if line.strip()]

        # Split the header line to extract column names
        header = lines[0].split(' | ')

        # Split the remaining lines to extract data
        data = [line.split(' | ') for line in lines[2:]]

        # Append the new rows to the existing CSV file
        filename = './sample_data/output.csv'
        with open(filename, 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerows(data)

        print(f"New rows appended to the CSV file '{filename}' successfully.")

    except Exception as e:
        print(f"Error: {e}. Skipping appending to the CSV file.")

# Example usage
text = "Header1 | Header2 | Header3\nData1 | Data2 | Data3"
appendToCharcterOutput(text)
```

```python
def genResponseFromGPT(pdf_text):

  question = f"""
Transform the following text about Marx into first person facts pretending to be this person:
Give the Reponse in a table format with title and fact columns formatted with newlines for each row. Each fact should be no longer then 400 characters long.

If the provided text contains no facts return a blank string: ''

If the text does contain facts, return them according to the following example:
base details | I am Karl Marx, born on May 5, 1818 in Trier Rhine province Prussia [Germany],
Death | I passed away on March 14\, 1883, in London\, England,

Here is the provided text:{pdf_text}"""

  # print(question)

  response = llm.predict(question)
  return response
```

```python
def parseBiography(file_path, page_num=0):
    with open(file_path, 'rb') as file:
        pdf = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(29 ,len(pdf.pages)):
            page = pdf.pages[page_num]
            text = page.extract_text()
            tableRes = genResponseFromGPT(text)
            appendToCharcterOutput(tableRes)
            print(page_num)
        return 'DONE'
```
